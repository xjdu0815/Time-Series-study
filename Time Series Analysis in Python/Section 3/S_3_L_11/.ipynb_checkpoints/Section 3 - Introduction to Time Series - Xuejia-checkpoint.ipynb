### Handling Missing Values

```python
df_comp.isna().sum()
```

At begining, there is no missing values. But now we have 8. This is because setting the frequency to 'business days' must have generated 8 dates, for which we have no data available. 

"fillna()" method:
* Front filling: assigns the value of the previous period
* Back filling: assigns the value for the next period.
* Assigning the same value: assign the average to all the missing values within the time series

```python
df_comp.spx =df_comp.spx.fillna(method ='ffill')
```

```python
df_comp.isna().sum()
```

```python
df_comp.ftse =df_comp.ftse.fillna(method ='bfill')
```

```python
df_comp.isna().sum()
```

```python
df_comp.dax =df_comp.dax.fillna(value =df_comp.dax.mean())
```

```python
df_comp.isna().sum()
```

You can fill the nikkei with any method you like. Here I choose average

```python
df_comp.nikkei =df_comp.nikkei.fillna(value = df_comp.dax.mean())
```

### White Noise 

white noise: a special type of time-series, where the data doesn't follow a pattern. <br> if you have mean is always 0, constant variance, no autocorrelation (no clear relationship between past and present values), then it's white noise.<br> white noise: a sequence of random data, where every value has a time-period associated with it.

```python
wn=np.random.normal(loc = df.market_value.mean(), scale = df.market_value.std(), size = len(df))
```

Try adding the white noise to the dataframe

```python
df['wn']=wn
```

```python
df.describe()
```

Normally distributed around the mean of the S&P. Since each one is generated individually, the average does not have to end up being the same.

```python
df.wn.plot(figsize=(20,5))
plt.title("white noise time-series", size =24)
plt.show()
```

```python
df.market_value.plot(figsize=(20,5))
plt.title("S&P Prices", size =24)
plt.ylim(0,2300) # change to the same axis
plt.show()
```

It's easy to see the difference

Constant mean, constant variance, and no autocorrelation in any period can be attributed to white noise.

# Random Walk

random walk: a special type of time-series, where values tend to persist over time and the differences between periods are simply white noise. The best estimator for today's value is yesterday's value. The best estimator for tomorrow's value is today's value. <br>What makes Random Walk different from white noise is the fact that the values aren’t a list of random numbers. The current value depends on the previous one.

```python
rw = pd.read_csv("RandWalk.csv")
rw.date = pd.to_datetime(rw.date, dayfirst=True)
rw.set_index("date", inplace =True)
rw = rw.asfreq("b")
```

```python
rw.describe()
```

```python
df['rw']=rw.price
```

```python
df.head()
```

```python
df.rw.plot(figsize=(20,5))
df.market_value.plot()
plt.title("Random Walk", size =24)
plt.show()
```

Small variations between consecutive time periods, cyclical increases and decreases in short periods of time.

Market efficiency: measures the level of difficulty in forecasting correct future values

# Stationarity

The mean is constant, the standard deviation is constant, no seasonality. White noise is stationarity.

# Seasonality

There are several ways to test seasonality, first is decomposition the sequency. We can split into 3 effects: Trend (pattern), seasonal (cyclical effects), residual (errors of predication)

"naive" decompostion has <br>1. additive: observed =trend+ seasonal+residual 
<br> 2.multiplicative:observed = trend*seasonal*residual

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
s_dec_additive =seasonal_decompose(df.market_value, model ='additive')
s_dec_additive.plot()
plt.show()
```

seasonal is rectangular, means when data oscilate back and forth, therefore no concrete cyclical pattern. Suggest no seasonality in data.

```python
s_dec_multiplicative =seasonal_decompose(df.market_value, model ='multiplicative')
s_dec_multiplicative.plot()
plt.show()
```

## Correlation between past and present values

Autocorrelation: there is a correlation between the values of the time series and a lagged version of itself. 

## The Autocorrelation Function (ACF)

ACF shows the relationship between past values and the current one.

```python
import statsmodels.graphics.tsaplots as sgt
```

```python
sgt.plot_acf(df.market_value, lags = 40, zero =False)
plt.title("ACF S&P", size =24)
plt.show()
```

The shadow area represents the siginicance,

```python
sgt.plot_acf(df.wn, lags = 40, zero =False)
plt.title("ACF White Noise", size =24)
plt.ylim(-0.03,0.03)
plt.show()
```

For white noise, no autocorrelation for any lag

```python
sgt.plot_acf(df.rw, lags = 40, zero =False)
plt.title("ACF random walk", size =24)
#plt.ylim(-0.01,0.1)
plt.show()
```

## Partial autocorrelation function (PACF)

```python
sgt.plot_pacf(df.market_value, lags =40, zero=False, method =('ols'))
#Order of Least Squares (OLS)
plt.title("PACF S&P", size=24)
plt.ylim(-0.25,1.1)
plt.show()
```

The data close to 0 means: not significantly different from 0, numeric values attached to them are not important. Being positive or negative is somewhat random without any lasting effects.

It cancels out all additional channels a previous period value affects the present one. ACF and PACF values for the first lags should be identical

```python
sgt.plot_pacf(df.wn, lags =40, zero=False, method =('ols'))
#Order of Least Squares (OLS)
plt.title("PACF WN", size=24)
plt.ylim(-0.03,0.03)
plt.show()
```
